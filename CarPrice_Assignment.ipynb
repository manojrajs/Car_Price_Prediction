## Importing and Understanding the data

#### Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Data Preparation

#### Conveting objects to categorical type

```python
## Removing columns which has not importance
cars = cars.drop(['car_ID','CarName'],axis=1)
```

```python
convert_dict = {'fueltype' : 'category', 'aspiration' : 'category',
       'doornumber': 'category', 'carbody' : 'category', 'drivewheel': 'category', 'enginelocation': 'category', 'enginetype' : 'category',
       'cylindernumber' : 'category','fuelsystem' : 'category'}
```

```python
cars = cars.astype(convert_dict)
```

```python
cars.info()
```

```python
cars.head()
```

#### Converting fueltype, aspiration, doornumber to 1 & 0
- fueltype = Gas - 0, Diesel - 1
- aspiration = std - 0, turbo - 1
- doornumber = two - 0, four - 1
- enginerlocation = front - 0, rear - 1

```python
# Map function works on dataframe and not series
```

```python
cars[['fueltype']] = cars[['fueltype']].apply(lambda x: x.map({'gas':0,'diesel':1}))
cars[['aspiration']] = cars[['aspiration']].apply(lambda x: x.map({'std':0,'turbo':1}))
cars[['doornumber']] = cars[['doornumber']].apply(lambda x: x.map({'two':0,'four':1}))
```

```python
cars[['enginelocation']] = cars[['enginelocation']].apply(lambda x: x.map({'front':0,'rear':1}))
```

```python
cars.head()
```

#### Creating dummy variables & Droping redundant variables

```python
cars.columns
```

```python
# if all zero then 'convertable' = 1 by default
status1 = pd.get_dummies(cars['carbody'],drop_first = True)
status1.head()
```

```python
# if all zeros then 4wd = 1 by default 
status2 = pd.get_dummies(cars['drivewheel'],drop_first = True)
status2.head()
```

```python
# if all zeros then dohc = 1 by default 
status3 = pd.get_dummies(cars['enginetype'],drop_first= True)
status3.head()
```

```python
# if all zeros then eight = 1 by default 
status4 = pd.get_dummies(cars['cylindernumber'],drop_first = True)
status4.head()
```

```python
# if all zeros then 1bbl = 1 by default 
status5 = pd.get_dummies(cars['fuelsystem'],drop_first = True)
status5.head()
```

```python
## Concating the dummy variables to cars
cars = pd.concat([cars,status1,status2,status3,status4,status5],axis=1)
cars.head()
```

```python
#removing the category columns
cars = cars.drop(['carbody','drivewheel','enginetype','cylindernumber','fuelsystem'],axis=1)
cars.head()
```

```python
cars.shape
```

## Splitting data into train and test

```python
from sklearn.model_selection import train_test_split 
```

```python
df_train,df_test = train_test_split(cars,train_size = 0.7,test_size=0.3,random_state=100)
```

### Rescaling the variables using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
# apply MinMaxScaler to all the non Yes/No and dummy variables
var = ['symboling','wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']
df_train[var] = scaler.fit_transform(df_train[var])
df_train.head()
```

```python
# confirming of MinMaxScaler
df_train.describe()
```

## Training & Building Model 

```python
# Splitting into x and y train
y_train = df_train.pop('price')
X_train = df_train
```

```python
X_train.head()
```

### Feature selection using RFE

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()

rfe = RFE(lm,n_features_to_select=30)
rfe = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train = X_train[col]
X_train.head()
```

## Using StatsModel for building the model

```python
import statsmodels.api as sm
```

```python
## Adding a constant
X_train_sm = sm.add_constant(X_train)

## Create a Model
lr = sm.OLS(y_train,X_train_sm)

## fit the model
lr_model = lr.fit()

## params
lr_model.summary()
```

### VIF
- High pvalue, High VIF
- High pvalue, Low VIF
- Low pvalue, High VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vf
```
